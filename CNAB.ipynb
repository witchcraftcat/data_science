```python
from google.colab import files
import pandas as pd


uploaded = files.upload()

```

```python
arquivo_planilha = next(iter(uploaded))
arquivo_planilha = list(uploaded.keys())[0]


def gerar_cnab(arquivo_planilha):
  df = pd.read_excel(arquivo_planilha, dtype = str)

  df["Banco"] = df["Banco"].astype(str)
  df["Agência"] = df["Agência"].astype(str).str.zfill(4)
  df["Conta"] = df["Conta"].astype(str).str.zfill(7)
  df["Valor"] = df["Valor"].astype(float)
  df["Valor"] = df["Valor"].apply(lambda x:f'{x:013.2f}')



  with open("arquivo_cnab_itau.txt", "w", encoding="utf-8", newline = "\r\n") as cnab:
    header = (
         "34100000         12345678901234567890EMPRESA LTDA       "
            "341BANCO ITAU SA     2025021100000108100000                         \r\n"
    )

    cnab.write(header)

  with open("arquivo_cnab_itau.txt", "a", encoding="utf-8", newline ="\r\n") as cnab:
    for index, row in df.iterrows():
      detalhe = (
          f"34100013A 000{row['CPF / CNPJ']:0<15} 000{row['Banco']}{row['Agência']:0>4}{row['Conta']:0>7} "
                f"{row['Nome do Favorecido'][:30]:<30}{float(row['Valor']):013.2f}{row['Data Pagamento']}                         \r\n"
            )
      print(f"Lendo linha {index}: {row.to_dict()}")
      cnab.write(detalhe)

  with open("arquivo_cnab_itau.txt", "a", encoding="utf-8", newline = "\r\n") as cnab:
    trailer = "34199999         000001                          \r\n"
    cnab.write(trailer)

  print("Arquivo gerado.")


gerar_cnab(arquivo_planilha)
```
